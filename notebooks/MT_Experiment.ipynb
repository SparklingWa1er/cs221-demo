```python
import os
import pandas as pd
import torch
import gc
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSeq2SeqLM,
    Seq2SeqTrainer,
    Seq2SeqTrainingArguments,
    DataCollatorForSeq2Seq
)
import warnings

warnings.filterwarnings("ignore")

# --- Cáº¤U HÃŒNH CÃC THÃ NGHIá»†M ---
EXPERIMENTS = [
    # 1. mT5-base
    {"name": "mt5_base_single",  "backbone": "google/mt5-base",  "mode": "single"},

    # 2. ViT5-base
    {"name": "vit5_base_single", "backbone": "google/mt5-large", "mode": "single"},
    {"name": "vit5_base_multi",  "backbone": "google/mt5-large", "mode": "multi"},

]

TRAIN_FILE = "dev_gemini-2.5-flash.csv"
OUTPUT_ROOT = "./experiment_results"

# Hyperparams (A100 Optimized)
EPOCHS = 3
LR = 3e-4
MAX_INPUT_LEN = 1024
MAX_TARGET_LEN = 256

def construct_prompt(row, mode):
    source = row['en']
    if mode == "single":
        cand = row['candidate_1']
        return f"Rewrite: Source: {source} | Cand: {cand}"
    else:
        c1, c2, c3 = row['candidate_1'], row['candidate_2'], row['candidate_3']
        return f"Rewrite: Source: {source} | C1: {c1} | C2: {c2} | C3: {c3}"

def run_training(exp_config):
    exp_name = exp_config["name"]
    model_name = exp_config["backbone"]
    mode = exp_config["mode"]
    output_dir = os.path.join(OUTPUT_ROOT, exp_name)

    # Check náº¿u Ä‘Ã£ train xong rá»“i thÃ¬ bá» qua
    if os.path.exists(os.path.join(output_dir, "final_model")):
        print(f"â© ÄÃ£ tÃ¬m tháº¥y model {exp_name}, bá» qua training.")
        return

    # Tá»± Ä‘á»™ng chá»‰nh Batch Size cho A100
    if "large" in model_name:
        batch_size = 2
    else:
        batch_size = 16

    print(f"\n{'='*40}")
    print(f"ğŸš€ TRAINING: {exp_name} | {model_name} | {mode.upper()}")
    print(f"âš™ï¸  Batch: {batch_size} | BF16: On")
    print(f"{'='*40}")

    # 1. Load Data
    df = pd.read_csv(TRAIN_FILE)
    df['input_text'] = df.apply(lambda x: construct_prompt(x, mode), axis=1)
    df['target_text'] = df['vi']

    dataset = Dataset.from_pandas(df[['input_text', 'target_text']])
    dataset = dataset.train_test_split(test_size=0.1, seed=42) # 90% train, 10% eval

    tokenizer = AutoTokenizer.from_pretrained(model_name)

    def preprocess(examples):
        inputs = examples['input_text']
        targets = examples['target_text']
        model_inputs = tokenizer(inputs, max_length=MAX_INPUT_LEN, truncation=True, padding="max_length")
        labels = tokenizer(targets, max_length=MAX_TARGET_LEN, truncation=True, padding="max_length")
        model_inputs["labels"] = [
            [(l if l != tokenizer.pad_token_id else -100) for l in label] for label in labels["input_ids"]
        ]
        return model_inputs

    tokenized_data = dataset.map(preprocess, batched=True)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

    # 2. Config Trainer
    args = Seq2SeqTrainingArguments(
        output_dir=output_dir,
        eval_strategy="epoch",
        save_strategy="no", # Chá»‰ lÆ°u model cuá»‘i cÃ¹ng
        learning_rate=LR,
        per_device_train_batch_size=batch_size,
        per_device_eval_batch_size=batch_size,
        num_train_epochs=EPOCHS,
        predict_with_generate=True,
        fp16=False,
        bf16=True,  # A100 dÃ¹ng cÃ¡i nÃ y Ä‘á»ƒ trá»‹ lá»—i NaN
        logging_steps=50,
        report_to="none"
    )

    trainer = Seq2SeqTrainer(
        model=model,
        args=args,
        train_dataset=tokenized_data["train"],
        eval_dataset=tokenized_data["test"],
        data_collator=DataCollatorForSeq2Seq(tokenizer, model=model),
        processing_class=tokenizer,
    )

    trainer.train()

    # 3. Save
    print(f"ğŸ’¾ Äang lÆ°u model vÃ o {output_dir}/final_model ...")
    trainer.save_model(f"{output_dir}/final_model")
    tokenizer.save_pretrained(f"{output_dir}/final_model")

    # 4. Clean up
    del model, trainer
    torch.cuda.empty_cache()
    gc.collect()

if __name__ == "__main__":
    if not os.path.exists(OUTPUT_ROOT): os.makedirs(OUTPUT_ROOT)
    for exp in EXPERIMENTS:
        try:
            run_training(exp)
        except Exception as e:
            print(f"âŒ Lá»—i khi train {exp['name']}: {e}")
```

```python
import os
import pandas as pd
import torch
import gc
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from sacrebleu.metrics import BLEU
from tqdm import tqdm

# --- Cáº¤U HÃŒNH GIá»NG Há»†T LÃšC TRAIN ---
EXPERIMENTS = [
    {"name": "mt5_base_single",  "backbone": "google/mt5-base",  "mode": "single"},
    {"name": "vit5_base_single", "backbone": "google/mt5-large", "mode": "single"},
    {"name": "vit5_base_multi",  "backbone": "google/mt5-large", "mode": "multi"},
]

TEST_FILE = "./data/vi_mt/test/Google/test_gemini-2.5-flash.csv"
OUTPUT_ROOT = "./experiment_results"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

def construct_prompt(row, mode):
    source = row['en']
    if mode == "single":
        return f"Rewrite: Source: {source} | Cand: {row['candidate_1']}"
    else:
        return f"Rewrite: Source: {source} | C1: {row['candidate_1']} | C2: {row['candidate_2']} | C3: {row['candidate_3']}"

def run_evaluation():
    results = []

    # 1. TÃ­nh Baseline trÆ°á»›c
    print("--- ğŸ“Š Äang tÃ­nh Baseline (Gemini 2.5) ---")
    df_test = pd.read_csv(TEST_FILE)
    bleu = BLEU(trg_lang="vi")
    refs = df_test['vi'].astype(str).tolist()

    # Baseline lÃ  candidate 2 (thÆ°á»ng tá»‘t nháº¥t) hoáº·c max cá»§a 3 candidates
    base_preds = df_test['candidate_2'].astype(str).tolist()
    baseline_score = bleu.corpus_score(base_preds, [refs]).score
    print(f"ğŸ”¹ Baseline BLEU: {baseline_score:.2f}")

    # 2. Duyá»‡t qua tá»«ng thÃ­ nghiá»‡m
    for exp in EXPERIMENTS:
        exp_name = exp["name"]
        model_path = f"{OUTPUT_ROOT}/{exp_name}/final_model"

        print(f"\nğŸ” Äang Ä‘Ã¡nh giÃ¡: {exp_name} ...")

        if not os.path.exists(model_path):
            print(f"âš ï¸ KhÃ´ng tÃ¬m tháº¥y model táº¡i {model_path}. Bá» qua.")
            continue

        try:
            # Load model
            tokenizer = AutoTokenizer.from_pretrained(model_path)
            model = AutoModelForSeq2SeqLM.from_pretrained(model_path).to(DEVICE)
            model.eval()

            # Inference
            generated_texts = []
            print("   -> Äang sinh vÄƒn báº£n...")

            # Batch inference thá»§ cÃ´ng Ä‘á»ƒ kiá»ƒm soÃ¡t
            for _, row in tqdm(df_test.iterrows(), total=len(df_test), desc=exp_name):
                prompt = construct_prompt(row, exp["mode"])
                inputs = tokenizer(prompt, return_tensors="pt", max_length=1024, truncation=True).to(DEVICE)

                with torch.no_grad():
                    outputs = model.generate(
                        **inputs,
                        max_length=256,
                        num_beams=4,
                        early_stopping=True
                    )
                generated_texts.append(tokenizer.decode(outputs[0], skip_special_tokens=True))

            # LÆ°u káº¿t quáº£ chi tiáº¿t
            df_res = df_test.copy()
            df_res['LMCOR_Output'] = generated_texts
            df_res.to_csv(f"{OUTPUT_ROOT}/{exp_name}/predictions.csv", index=False)

            # TÃ­nh Ä‘iá»ƒm
            score = bleu.corpus_score(generated_texts, [refs]).score

            results.append({
                "Model": exp["backbone"],
                "Mode": exp["mode"].upper(),
                "BLEU": round(score, 2),
                "Gain": round(score - baseline_score, 2)
            })

            print(f"   âœ… Káº¿t quáº£: {score:.2f} (Gain: {score - baseline_score:.2f})")

            # Dá»n dáº¹p
            del model, tokenizer
            torch.cuda.empty_cache()
            gc.collect()

        except Exception as e:
            print(f"âŒ Lá»—i Ä‘Ã¡nh giÃ¡ {exp_name}: {e}")

    # 3. Tá»•ng há»£p bÃ¡o cÃ¡o
    if results:
        print("\n" + "="*60)
        print("ğŸ† Báº¢NG Xáº¾P Háº NG Káº¾T QUáº¢ THá»°C NGHIá»†M ğŸ†")
        print("="*60)
        df_final = pd.DataFrame(results).sort_values(by="BLEU", ascending=False)
        print(df_final.to_string(index=False))
        df_final.to_csv("FINAL_BENCHMARK_REPORT.csv", index=False)
    else:
        print("KhÃ´ng cÃ³ káº¿t quáº£ nÃ o.")

if __name__ == "__main__":
    run_evaluation()
```

```python
import os
import pandas as pd
import torch
import gc
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from sacrebleu.metrics import BLEU
from tqdm import tqdm
from comet import download_model, load_from_checkpoint
import warnings

warnings.filterwarnings("ignore")

# --- Cáº¤U HÃŒNH ---
EXPERIMENTS = [
    {"name": "mt5_base_single",  "backbone": "google/mt5-base",   "mode": "single"},
    {"name": "vit5_base_single", "backbone": "VietAI/vit5-base",  "mode": "single"},
    {"name": "vit5_base_multi",  "backbone": "VietAI/vit5-base",  "mode": "multi"},
]

TEST_ROOT_DIR = "./data/vi_mt/test"
OUTPUT_ROOT = "./experiment_results"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# !!! THÃŠM Má»šI: Tá»· lá»‡ dá»¯ liá»‡u muá»‘n cháº¡y thá»­ (0.05 = 5%) !!!
DEBUG_RATIO = 1
# Náº¿u muá»‘n cháº¡y Ã­t nháº¥t 5 dÃ²ng Ä‘á»ƒ trÃ¡nh lá»—i metric khi file quÃ¡ nhá»:
MIN_SAMPLES = 5

# --- LOAD COMET MODEL ---
print("â³ Äang load model COMET (wmt22-comet-da)...")
try:
    comet_path = download_model("Unbabel/wmt22-comet-da")
    comet_model = load_from_checkpoint(comet_path)
    comet_model.eval().to(DEVICE)
    HAS_COMET = True
    print("âœ… ÄÃ£ load COMET thÃ nh cÃ´ng.")
except Exception as e:
    print(f"âš ï¸ Lá»—i load COMET: {e}")
    HAS_COMET = False

def construct_prompt(row, mode):
    source = row['en']
    if mode == "single":
        return f"Rewrite: Source: {source} | Cand: {row['candidate_1']}"
    else:
        return f"Rewrite: Source: {source} | C1: {row['candidate_1']} | C2: {row['candidate_2']} | C3: {row['candidate_3']}"

def find_all_test_files(root_dir):
    csv_files = []
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith(".csv"):
                csv_files.append(os.path.join(root, file))
    return csv_files

def run_evaluation():
    all_scores = []

    test_files = find_all_test_files(TEST_ROOT_DIR)
    print(f"ğŸ“‚ TÃ¬m tháº¥y {len(test_files)} file test trong {TEST_ROOT_DIR}")

    # --- VÃ’NG Láº¶P 1: DUYá»†T QUA Tá»ªNG MODEL ---
    for exp in EXPERIMENTS:
        exp_name = exp["name"]
        model_path = os.path.join(OUTPUT_ROOT, exp_name, "final_model")

        print(f"\n{'='*60}")
        print(f"ğŸš€ Äang Ä‘Ã¡nh giÃ¡ Model: {exp_name}")
        print(f"{'='*60}")

        if not os.path.exists(model_path):
            print(f"âš ï¸ KhÃ´ng tÃ¬m tháº¥y model táº¡i {model_path}. Bá» qua.")
            continue

        try:
            tokenizer = AutoTokenizer.from_pretrained(model_path)
            model = AutoModelForSeq2SeqLM.from_pretrained(model_path).to(DEVICE)
            model.eval()

            # --- VÃ’NG Láº¶P 2: CHáº Y TRÃŠN Tá»ªNG FILE TEST ---
            for file_path in test_files:
                file_name = os.path.basename(file_path)
                print(f"   ğŸ“„ File: {file_name}")

                try:
                    df_test = pd.read_csv(file_path)

                    # --- [NEW] Cáº®T GIáº¢M Dá»® LIá»†U Äá»‚ TEST PIPELINE ---
                    if DEBUG_RATIO < 1.0:
                        original_len = len(df_test)
                        # Láº¥y 5% hoáº·c tá»‘i thiá»ƒu MIN_SAMPLES dÃ²ng
                        n_samples = max(MIN_SAMPLES, int(original_len * DEBUG_RATIO))
                        # KhÃ´ng láº¥y quÃ¡ sá»‘ lÆ°á»£ng dÃ²ng thá»±c táº¿
                        n_samples = min(n_samples, original_len)

                        df_test = df_test.sample(n=n_samples, random_state=42).reset_index(drop=True)
                        print(f"      âœ‚ï¸ [DEBUG MODE] Äang cháº¡y trÃªn {n_samples}/{original_len} dÃ²ng ({DEBUG_RATIO*100}%)")
                    # -----------------------------------------------

                    # 1. Sinh vÄƒn báº£n (Inference)
                    generated_texts = []
                    for _, row in tqdm(df_test.iterrows(), total=len(df_test), desc=f"      Gen ({file_name[:15]}...)", leave=False):
                        prompt = construct_prompt(row, exp["mode"])
                        inputs = tokenizer(prompt, return_tensors="pt", max_length=1024, truncation=True).to(DEVICE)

                        with torch.no_grad():
                            outputs = model.generate(
                                **inputs,
                                max_length=256,
                                num_beams=4,
                                early_stopping=True
                            )
                        generated_texts.append(tokenizer.decode(outputs[0], skip_special_tokens=True))

                    save_dir = os.path.join(OUTPUT_ROOT, exp_name, "predictions_debug" if DEBUG_RATIO < 1.0 else "predictions")
                    os.makedirs(save_dir, exist_ok=True)

                    df_res = df_test.copy()
                    df_res['LMCOR_Output'] = generated_texts
                    pred_file_path = os.path.join(save_dir, f"pred_{file_name}")
                    df_res.to_csv(pred_file_path, index=False)

                    # 2. TÃ­nh BLEU
                    preds = generated_texts
                    refs = df_test['vi'].astype(str).tolist()
                    srcs = df_test['en'].astype(str).tolist()

                    if len(preds) > 0:
                        bleu = BLEU(trg_lang="vi")
                        bleu_score = bleu.corpus_score(preds, [refs]).score
                    else:
                        bleu_score = 0.0

                    # 3. TÃ­nh COMET
                    comet_score = 0.0
                    if HAS_COMET and len(preds) > 0:
                        comet_data = [{"src": s, "mt": p, "ref": r} for s, p, r in zip(srcs, preds, refs)]
                        # Giáº£m batch size náº¿u data Ã­t
                        batch_size_comet = min(64, len(comet_data))
                        comet_out = comet_model.predict(comet_data, batch_size=batch_size_comet, gpus=1, progress_bar=False)
                        comet_score = comet_out.system_score * 100

                    # 4. Ghi nháº­n káº¿t quáº£
                    all_scores.append({
                        "Experiment": exp_name,
                        "Backbone": exp["backbone"],
                        "Mode": exp["mode"],
                        "Test_File": file_name,
                        "BLEU": round(bleu_score, 2),
                        "COMET": round(comet_score, 2),
                        "Data_Size": len(df_test) # Note láº¡i size Ä‘á»ƒ biáº¿t lÃ  báº£n test
                    })

                    print(f"      âœ… BLEU: {bleu_score:.2f} | COMET: {comet_score:.2f}")

                except Exception as e:
                    print(f"      âŒ Lá»—i xá»­ lÃ½ file {file_name}: {e}")
                    import traceback
                    traceback.print_exc()

            del model, tokenizer
            torch.cuda.empty_cache()
            gc.collect()

        except Exception as e:
            print(f"âŒ Lá»—i load model {exp_name}: {e}")

    # --- LÆ¯U BÃO CÃO Tá»”NG Há»¢P ---
    if all_scores:
        print("\n" + "="*60)
        print("ğŸ’¾ ÄANG LÆ¯U Káº¾T QUáº¢ Tá»”NG Há»¢P...")
        df_final = pd.DataFrame(all_scores)
        df_final = df_final.sort_values(by=["Test_File", "BLEU"], ascending=[True, False])

        print(df_final.to_string(index=False))

        save_name = "DEBUG_EVAL.csv" if DEBUG_RATIO < 1.0 else "FULL_EVAL_BLEU_COMET.csv"
        df_final.to_csv(save_name, index=False)
        print(f"\nâœ… ÄÃ£ lÆ°u file táº¡i: {save_name}")
    else:
        print("âš ï¸ KhÃ´ng cÃ³ káº¿t quáº£ nÃ o Ä‘Æ°á»£c ghi nháº­n.")

if __name__ == "__main__":
    run_evaluation()
```
